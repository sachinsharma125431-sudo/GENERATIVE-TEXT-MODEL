```python
import markovify  # Library used to generate text using Markov chains

# -------------------- Training Data --------------------

# Sample dataset used to train the Markov model
data = """
Artificial Intelligence is changing the world rapidly.
AI is used in healthcare to detect diseases early.
Machine learning helps systems learn from data automatically.
Deep learning is used in image recognition and speech processing.
Technology is improving day by day with AI advancements.
AI is also used in finance, automation, and robotics.
Many companies are investing in artificial intelligence.
AI can solve complex real world problems efficiently.
"""

# Create a Markov model based on the given text
model = markovify.Text(data)

# Take topic input from user
topic = input("Enter any topic: ")

print("\nGenerated Text:\n")

# Initialize counters
count = 0      # Number of valid sentences generated
attempt = 0    # Number of attempts made to generate sentences

# -------------------- Text Generation Loop --------------------

# Try generating up to 3 sentences, with maximum 10 attempts
while count < 3 and attempt < 10:
    
    # Generate a sentence using the Markov model
    sentence = model.make_sentence()
    
    # If a valid sentence is generated
    if sentence:
        # Print sentence with the user-provided topic added in front
        print("- " + topic + " " + sentence)
        count += 1  # Increase successful sentence count
    
    attempt += 1  # Increase attempt count

# -------------------- Fallback Output --------------------

# If no valid sentences were generated, print default lines
if count == 0:
    print("- " + topic + " AI is growing rapidly in today's world.")
    print("- " + topic + " Machine learning is very useful in real life.")
```
